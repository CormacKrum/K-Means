```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(centers=6,random_state=42)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
model = KMeans(n_clusters=6,max_iter=1)

model.fit(X)
```

```python
cluster_centers = model.cluster_centers_
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.scatter(cluster_centers[:,0],cluster_centers[:,1],c='0',s=100)
plt.show()
```

```python
inertia_output=[]
for i in range(1,10):
    model = KMeans(n_clusters=i,max_iter=1)
    model.fit(X)
    inertia_output.append(model.inertia_)
```

```python
steps = np.arange(1,10)
```

```python
plt.plot(steps,inertia_output)
```
