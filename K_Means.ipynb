# K-Means Clustering


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,y=make_blobs(centers=6,random_state=42)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=6,n_init=1,max_iter=1)
model.fit(X)
#model.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=model.labels_)
for center in model.cluster_centers_:
    plt.scatter(center[0],center[1],s=80)
```

```python
model.cluster_centers_
```

```python
model.inertia_
```

```python
step=np.arange(1,10)
out=[]
for i in step:
    model=KMeans(n_init=5,n_clusters=i)
    model.fit(X)
    out.append(model.inertia_)
```

```python
out
```

```python
plt.plot(step,out)
```
